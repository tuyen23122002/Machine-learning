```python
class Manufacturer:
  def __init__(self,id,location):
    self._id = id
    self._location = location
  def describe(self):
    print(f"Id:{self._id} - Location:{self._location}")
```

```python
class Device(Manufacturer):
  def __init__(self,name,price,id,location):
    self._name = name
    self._price = price
    self._location = location
    self._id = id
  def describe(self):
    print(f"Name:{self._name} - Price:{self._price} - Id:{self._id} - Location:{self._location}")
```

```python
device1 = Device("mouse",25,9734,"VietNam")
device1.describe()
```

```python
class Mystack:
  def __init__(self,capacity):
    self.__capacity = capacity
    self.__stack = list()

  def isfull(self):
    return len(self.__stack) == self.__capacity

  def isEmpty(self):
    return len(self.__stack) == None

  def pop(self):
    return self.__stack.pop()

  def push(self,x):
    self.__stack.append(x)

  def top(self):
    return self.__stack[-1]

  def describe(self):
    for i in self.__stack:
      print(i)
```

```python
mystack = Mystack(5)
mystack.push(1)
mystack.push(2)
mystack.push(3)
mystack.push(3)
mystack.push(3)

print(mystack.top())

```
