```python
def computedistance(datapoint1,datapoint2):
  result =0
  for i in range (4):
    result += (float(datapoint1[i]) - float(datapoint2[i]))**2

  return math.sqrt(result)
```

```python
def computeKnearestNeighbor(trainingSet, item, k):
    distances = []
    for datapoint in trainingSet:
        distances.append({
            "label": datapoint[-1],
            "value": computedistance(item, datapoint)
        })

    # sắp xếp các khoảng cách theo thứ tự tăng dần
    distances.sort(key = lambda x: x["value"])

    # lấy nhãn của k khoảng cách nhỏ nhất
    labels = [item["label"] for item in distances]

    # trả về k nhãn nhỏ nhất
    return labels[:k]
```

```python
def Votingdistances(list):
    element_count = {}
    for i in list :
        if i in element_count:
            element_count[i] += 1
        else:
            element_count[i] = 1

    sort = sorted(element_count.items(),key = lambda x: x[1], reverse = True)
    return sort[0][0]
```

```python
k =9
for item in testingSet:
  knn = computeKnearestNeighbor(trainingSet,item,k)
  result = Votingdistances(knn)

  print("label :{} -> predict: {}".format(item[-1],result))
```
